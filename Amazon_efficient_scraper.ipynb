```python
# Function to extract Product Title hoo
def get_title(soup):
    try:
        # Outer Tag Object
        title = soup.find("span", attrs={"id": 'productTitle'})

        # Inner NavigatableString Object
        title_value = title.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = "Null"

    return title_string


# Function to extract Product Price
def get_price(soup):
    try:
        price = soup.find("span", attrs={'class': 'a-price-whole'}).string.strip()

    except AttributeError:

        try:
            # If there is some deal price
            price = soup.find("span", attrs={'class': 'a-offscreen'}).string.strip()

        except:
            price = "Null"

    return price


# Function to extract Product Rating
def get_rating(soup):
    try:
        rating = soup.find("i", attrs={'class': 'a-icon a-icon-star a-star-4-5'}).string.strip()

    except AttributeError:
        try:
            rating = soup.find("span", attrs={'class': 'a-icon-alt'}).string.strip()
        except:
            rating = "Null"

    return rating


# Function to extract Number of User Reviews
def get_review_count(soup):
    try:
        review_count = soup.find("span", attrs={'id': 'acrCustomerReviewText'}).string.strip()

    except AttributeError:
        review_count = "Null"

    return review_count


# Function to extract Availability Status
def get_availability(soup):
    try:
        available = soup.find("td", attrs={'class': 'a-size-base prodDetAttrValue'}).string.strip()
    #         available = available.find("span").string.strip()

    except AttributeError:
        available = "Not Available"

    return available

# Function to extract ASIN OF Listing
def get_ASIN(soup):
    try:
        ASIN = soup.find("td", attrs={'class': 'a-size-base prodDetAttrValue'}).string.strip()
    #         available = available.find("span").string.strip()

    except AttributeError:
        ASIN = "Not Available"

    return ASIN

# Function to extract date of Availability on Amazon catalogue
def DOA(soup):
    try:
        l1 = soup.find('table', attrs={"id": "productDetails_detailBullets_sections1"})
        f = 0
        for i in l1.find_all('th'):
            if i.string.strip() == "Date First Available":
                break
            else:
                f += 1
        return l1.find_all('td')[f].string.strip()
    except:
        return "Null"


def asin(soup):
    try:
        l1 = soup.find('table', attrs={"id": "productDetails_detailBullets_sections1"})
        return l1.find('td').string.strip()
    except:
        return "Null"

# Function to extract number of pages of products of seller
def find_pages(soup):
    try:
        try:
            no_pages = soup.find("span", attrs={'class': 's-pagination-item s-pagination-disabled'}).string.strip()
            return no_pages
        except:
            k = soup.find_all(attrs={'class': 's-pagination-item s-pagination-button'})
            return max([x.string.strip() for x in k])
    except:
        return "null"
```

```python
#Function to extract data from product page
#this function is also used for our concurrent process execution
def get_details(url):
    HEADERS = ({'User-Agent': '', 'Accept-Language': 'en-US, en;q=0.5'})
    response = requests.get(url, headers=HEADERS, proxies=get_proxy())
    new_soup = BeautifulSoup(response.content, 'lxml')

    d = {"title": "", "price": "", "rating": "", "reviews": "", "availability": "", 'Date First Available': "",
         'ASIN': ""}

    d['title'] += (get_title(new_soup))
    d['price'] += (get_price(new_soup))
    d['rating'] += (get_rating(new_soup))
    d['reviews'] += (get_review_count(new_soup))
    d['Date First Available'] += (DOA(new_soup))
    d['ASIN'] += (asin(new_soup))

    return d
```

```python
if __name__ == '__main__':

    # add your user agent
    HEADERS = ({'User-Agent': '', 'Accept-Language': 'en-US, en;q=0.5'})

    # The webpage URL
    # URL = input("Enter Url of Amazon seller storefront :")
    URL = "https://www.amazon.in/s?me=A2AL6IVND0I91F&marketplaceID=A21TJRUUN4KGV"
    merchant_id = URL[27:41]

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS,proxies=get_proxy())

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "lxml")

    # extract no. of pages
    no_pages = find_pages(soup)
    print("Total no. of pages will be ", no_pages)
    print("Want to enter manually no. of pages ?(Y/N)")
    choice = input()
    if choice in ["Y", "y"]:
        no_pages = int(input("Enter no. of pages :"))
    # file_name = input("enter file name :")

    start_time = time.time()
    # scrape the whole product links in a particular page
    for i in range(1, int(no_pages) + 1):
        NEW_URL = f"https://www.amazon.in/s?i=merchant-items&me={merchant_id}&page={i}&marketplaceID=A21TJRUUN4KGV"

        webpage = requests.get(NEW_URL, headers=HEADERS,proxies=get_proxy())

        # Soup Object containing all data
        soup = BeautifulSoup(webpage.content, "lxml")

        # Fetch links as List of Tag Objects
        links = soup.find_all("a", attrs={'class': 'a-link-normal s-no-outline'})

        # Store the links of all product in page
        links_list = ["https://www.amazon.in"+link.get('href') for link in links]

        with ThreadPoolExecutor(max_workers=20) as executor:
            # Create thread pool for scraping multiple product pages concurrently
            results = list(executor.map(get_details, links_list))

    end_time = time.time()
    execution_time = end_time - start_time
    print(f"Total execution time for {no_pages} pages is", round(execution_time, 3), " seconds")
```
